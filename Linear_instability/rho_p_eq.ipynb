## Linear stability analysis for the density and polarity equations
The equations for $\rho_S$, $\mathbf{p}_S$ of species $S$ read
$$
\begin{align*}
	\partial_t \rho_S &= -\nabla \cdot (v_S \mathbf{p}_S) \\
	\partial_t \mathbf{p}_S &= \frac{v_S}{16D_r} \Delta (v_S\mathbf{p}_S) - \frac{1}{2} \nabla (v_S \rho_S) - D_r \mathbf{p}_S,
\end{align*}
$$
where $v_S$ is determined by local densities of both species as
$$
	v_S = \bar{v}_{S} \left [1
	+ \eta_{S A} \frac{\rho_A (\mathbf{r}) - \bar{\rho}_{A}}{\rho_0}
	+ \eta_{S B} \frac{\rho_B (\mathbf{r}) - \bar{\rho}_{B}}{\rho_0} 
	\right ].
$$
Around the homogenous phase $(\rho_S=\rho_0, \mathbf{p}_S=0)$, the Fourier modes of fluctuations evolve as
$$
\partial_t \begin{pmatrix}
		\delta \rho_A \\ p_A \\ \delta \rho_B \\ p_B
	\end{pmatrix} = v_0
	\begin{pmatrix}
	0 & -iq & 0 & 0 \\
	-\frac{iq}{2}(1+\eta_{AA}) & -D & -\frac{iq}{2}\eta_{AB} & 0 \\
	0 & 0 & 0 & -iq \\
	-\frac{iq}{2}\eta_{BA} & 0 & -\frac{iq}{2}(1+\eta_{BB}) & -D
	\end{pmatrix} \begin{pmatrix}
		\delta \rho_A \\ p_A \\ \delta \rho_B \\ p_B
	\end{pmatrix} = v_0 M \begin{pmatrix}
		\delta \rho_A \\ p_A \\ \delta \rho_B \\ p_B
	\end{pmatrix},
$$

where $D=\frac{D_r}{v_0}+\frac{v_0}{16D_r}q^2$.
By setting $\varepsilon_{XY} = -\frac{iq}{2} (\delta_{X,Y}+\eta_{XY})$, the matrix $M$ can be simplified as:
$$
	M = \begin{pmatrix}
		0 & -iq & 0 & 0 \\
		\varepsilon_{AA} & -D & \varepsilon_{AB} & 0 \\
		0 & 0 & 0 & -iq \\
		\varepsilon_{BA} & 0 & \varepsilon_{BB} & -D
	\end{pmatrix}.
$$

To get the eigenvalues of the matrix $M$, we calculate the determinant
$$
\begin{align*}
	|M-\lambda I| &= \left |
	\begin{matrix}
		-\lambda & -iq & 0 & 0 \\
		\varepsilon_{AA} & -D-\lambda & \varepsilon_{AB} & 0 \\
		0 & 0 & -\lambda & -iq \\
		\varepsilon_{BA} & 0 & \varepsilon_{BB} & -D - \lambda
	\end{matrix} \right | \\
	&= \left [\lambda^2 + D \lambda + \frac{iq(\varepsilon_{AA}+\varepsilon_{BB})}{2} \right ]^2 +
	\left [\frac{(\varepsilon_{AA}-\varepsilon_{BB})^2}{4} + \varepsilon_{AB}\varepsilon_{BA}\right ]q^2 \\
	&= \left [\lambda^2 - D\lambda + \frac{(2+\eta_{AA}+\eta_{BB})q^2}{4} \right ]^2 -
	\left [\frac{(\eta_{AA}-\eta_{BB})^2}{16} + \frac{\eta_{AB}\eta_{BA}}{4} \right ] q^4,
\end{align*}
$$

Letting $|M-\lambda I|=0$, $\lambda$ should obey:
$$
	\lambda^2 + D\lambda + \frac
		{i(\varepsilon_{AA}+\varepsilon_{BB})\pm \sqrt{-(\varepsilon_{AA}-\varepsilon_{BB})^2 - 4\varepsilon_{AB}\varepsilon_{BA}}}
		{2}q = 0,
$$

which leads to
$$
\begin{align*}
	\lambda_{\pm\pm} &=
	\frac
	{-D \pm \sqrt{D^2 - 2 \left [i(\varepsilon_{AA} + \varepsilon_{BB}) \pm \sqrt{-(\varepsilon_{AA}-\varepsilon_{BB})^2 - 4\varepsilon_{AB}\varepsilon_{BA}}
	\right ]q}}
	{2},\\
	&= \frac{-D\pm\sqrt{D^2 - 2 \left [
		\frac{2+\eta_{AA}+\eta_{BB}}{2} \pm \sqrt{\frac{(\eta_{AA} - \eta_{BB})^2}{4} + \eta_{AB}\eta_{BA}}
		\right ] q^2}}{2}, \\
	&= 	\frac
	{-D\pm\sqrt{D^2-2\mu q^2 \pm 2\sqrt{\Delta}q^2}}
	{2},
\end{align*}
$$
where we have define
$$
\begin{align*}
\mu &\equiv \frac{2 + \eta_{AA} + \eta_{BB}}{2}, \\
\Delta &\equiv \frac{(\eta_{AA}-\eta_{BB})^2}{4} + \eta_{AB} \eta_{BA}.
\end{align*}
$$

1. When
$\Delta \geq 0$, it's impossible to have $\Re(\lambda)>0$ and $\Im(\lambda)\neq 0$ simultaneously, so we don't expect to find dynamic patterns here. Instead, in this case we can find static patterns when
$$
	\mu \leq 0\quad \mathrm{or} \quad \frac{\mu^2}{\Delta} > 1.
$$
2. When $\Delta < 0$, the expression of $\lambda$ becomes
$$\lambda_{\pm\pm}=\frac{-D\pm \sqrt{D^2 - 2\mu q^2 \pm 2i \sqrt{-\Delta} q^2}}{2},$$
and $\lambda$ always has a nonzero imaginary part. For the real part, it's obvious that $\Re(\lambda_{++})=\Re(\lambda_{+-})\geq \Re(\lambda_{-+})=\Re(\lambda_{--})$, so we need only consider $\Re(\lambda_{++})$, by inspecting which we find:

  * If $\mu \leq 0$, $\Re(\lambda_{+\pm}) > 0$ for $q>0$.
  * If $\mu > 0$, let $(a+ib)^2=D^2-2\mu q^2 + 2i\sqrt{-\Delta}q^2$, then

$$a^2-b^2=D^2 - 2\mu q^2, \quad ab=\sqrt{-\Delta}q^2,$$

which leads to
$$a^4 - (D^2-2\mu q^2)a^2 + \Delta q^4 = 0.$$

Thus
$$ a^2 = \frac{D^2-2\mu q^2 \pm \sqrt{(D^2-2\mu q^2)^2-4\Delta q^4}}{2}
$$

To let $\lambda$ have positive real part, we need $a^2>D^2$, which requires
$$\pm \sqrt{(D^2-2\mu q^2)^2-4\Delta q^4} > D^2 + 2\mu q^2,
$$
leading to
$$-q^2\Delta> 2\mu D^2
$$

When $\mu <=0$, this inequality is true for all $q>0$.


When $\mu >0$, substituting $D=\frac{D_r}{v_0}+\frac{v_0}{16D_r}q^2$ into this equation, we get
$$ \frac{\mu v_0^2}{16^2D_r^2}q^4 + (\frac{\mu}{8}-\nu^2)q^2 + \frac{\mu D_r^2}{v_0^2} < 0
$$

The condition for the inequality to be true is
$$
    (\frac{\mu}{8}-\nu^2)^2- (\frac{\mu}{8})^2 > 0,
$$
which requires
$$ \nu^2 > \frac{\mu}{4}$$

When $\nu^2 > \frac{\mu}{4}$, the inequality is true when $q_- < q < q_+$, where $q_\pm >0$ satisfy
$$
	q^2_\pm = \frac{1}{2} \left (\frac{16D_r}{v_0}\right )^2 \left [
	\left (\frac{\nu^2}{\mu}-\frac{1}{8}\right ) \pm \sqrt{\left (\frac{\nu^2}{\mu}-\frac{1}{8}\right )^2-\left (\frac{1}{8}\right )^2}
	\right ].
$$
In our system, there is a minimal $q^*$ set by the microscopic details such as interaction range. To get positive $\Re(\lambda_{+\pm})$ at valid $q$, one need $q^* > q_-$, which leads to the second inequation:
$$
	\frac{D_r}{v_0} < \frac{D_r^*}{v_0} =
	\frac{\sqrt{2}}{16} \left [
	\left (\frac{\nu^2}{\mu}-\frac{1}{8}\right ) - \sqrt{\left (\frac{\nu^2}{\mu}-\frac{1}{8}\right )^2-\left (\frac{1}{8}\right )^2}
	\right ]^{-\frac{1}{2}} q^*.
$$

```python
def get_Dr_threshold(eta, eta_AB, eta_BA, v0=1, q_thresh=np.pi*2):
    mu = 2 + 2 * eta
    nu = (-eta_AB * eta_BA) ** 0.5
    a = nu**2 / mu - 1/8
    Dr_thresh = v0 * np.sqrt(2) / 16 * (a - np.sqrt(a**2 - 1/64)) ** (-0.5) * q_thresh
    return Dr_thresh
```

Denoting $b=\frac{\nu^2}{\mu}-\frac{1}{8}$, at fixed $D_r$, when $q^*>q_-$, we have
$$a > b - \sqrt{b^2 - \left(\frac{1}{8}\right)^2},$$
where we have defined $a=2\left(\frac{v_0}{16D_r}\right)^2 {q^*}^2$.
That is to say,
$$b-a < \sqrt{b^2 - \left(\frac{1}{8}\right)^2}$$
Given that 

$$b < a \quad \mathrm{or}\quad b>\frac{a^2 + \left(\frac{1}{8}\right)^2} {2a}.$$

If $\frac{a^2 + \left(\frac{1}{8}\right)^2}{}$


$$\frac{\nu^2}{\mu}>\frac{(a+\frac{1}{8})^2}{2a}\equiv c$$

If setting $\mu = 2 + 2\eta$, $\nu^2=\alpha^2 - \eta^2$, the threshold for $\alpha$ is
$${\alpha^*}^2>\eta^2 + 2c(1+\eta).$$

If setting $\nu^2=\eta_{AB}^2$, the threshold for $\eta_{AB}$ is
$${\eta_{AB}^*}^2>2c(1+\eta),$$
where $\eta > -1$.

```python
def cal_alpha_threshold(eta, Dr, qc=np.pi*2, v0=1):
    a = 2 * (v0/(16 * Dr))** 2 * qc ** 2
    c = (a + 1/8) ** 2 / (2 * a)
    return np.sqrt(eta ** 2 + 2 * c * (1+eta))

def cal_etaAB_threshold(eta, Dr, qc=np.pi*2, v0=1):
    a = 2 * (v0/(16 * Dr))** 2 * qc ** 2
    c = (a + 1/8) ** 2 / (2 * a)
    return np.sqrt(2 * c * (1+eta))
```

## Plot figures
### Plot $\Re(\lambda)$ for various parameters

Define the function to calculate $\Re(\lambda_{++})$, where $\lambda$ is
$$\lambda_{\pm\pm}=\frac{-D\pm \sqrt{D^2 - \mu q^2 \pm 2i \nu q^2}}{2},$$
and $D=\frac{D_r}{v_0}+\frac{v_0}{16D_r}q^2$,
$\mu = 2 + \eta_{AA} + \eta_{BB}$,
$ \nu = \sqrt{-\frac{(\eta_{AA}- \eta_{BB})^2}{4} - \eta_{AB}\eta_{BA}}$.


```python
def cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q, v0=1):
    qq = q ** 2
    D = Dr / v0 + v0 / (16 * Dr) * q ** 2
    mu = 2 + eta + eta
    nu = np.sqrt(-eta_AB * eta_BA)
    return 0.5 * (-D + np.sqrt(D*D - mu * qq + 2j * nu * qq))



def plot_lambda_real():
    q = np.logspace(-2, 3, 1000)

    eta = 0
    Dr = 2
    eta_AB = 1
    eta_BA = - eta_AB

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(9, 6), constrained_layout=True)
    
    # varied alpha at eta=0 and Dr=0.1
    eta = -0.3
    Dr = 3
    for alpha2 in [0.2, 0.5, 1, 1.5]:
        eta_AB = np.sqrt(alpha2)
        eta_BA = -eta_AB
        y = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q).real
        ax1.plot(q, y, label=r"$%.1f$" % alpha2)
    ax1.set_title("(a) $\eta=%g,D_r=%g$" % (eta, Dr), fontsize="large")
    
    # varied Dr at eta=0 and alpha=1
    eta = 0
    eta_AB = 1
    eta_BA = -eta_AB
    Dc = get_Dr_threshold(eta, eta_AB, eta_BA)
    for Dr in [1, Dc, 10]:
        y = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q).real
        ax2.plot(q, y, label=r"$%.2f$" % Dr)
    ax2.set_title("(b) $\eta=%g,\eta_{AB}=-\eta_{BA}=%g$" % (eta, eta_AB), fontsize="large")

    # varied eta at alpha=1 and Dr=0.1
    eta_AB = 1
    eta_BA = -1
    Dr = 0.1
    for eta in [0, -0.5, -1, -2]:
        y = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q).real
        ax3.plot(q, y, label=r"$%g$" % eta)
    ax1.axhline(0, linestyle="dashed", c="k")
    ax2.axhline(0, linestyle="dashed", c="k")
    ax3.axhline(0, linestyle="dashed", c="k")
    ax3.set_title("(c) $D_r=%g,\eta_{AB}=-\eta_{BA}=%g$" % (Dr, eta_AB), fontsize="large")

    ax1.axvline(np.pi * 2, linestyle="dotted", c="tab:red")
    ax2.axvline(np.pi * 2, linestyle="dotted", c="tab:red")
    ax3.axvline(np.pi * 2, linestyle="dotted", c="tab:red")

    ax1.set_xscale("log")
    ax2.set_xscale("log")
    ax3.set_xscale("log")

    ax1.set_xlim(q[0], q[-1])
    ax2.set_xlim(q[0], q[-1])
    ax3.set_xlim(q[0], q[-1])
    
    ax2.set_ylim(-10, 5)

    ax1.set_xlabel(r"$q$", fontsize="x-large")
    ax2.set_xlabel(r"$q$", fontsize="x-large")
    ax3.set_xlabel(r"$q$", fontsize="x-large")
    ax1.set_ylabel(r"$q\Re(\lambda_{++})$", fontsize="x-large")

    ax1.legend(title="$\eta_{AB}^2=$", fontsize="large")
    ax2.legend(title="$D_r=$", fontsize="large")
    ax3.legend(title="$\eta=$", fontsize="large")

    plt.show()
    # plt.savefig("Re_lambda.pdf")
    plt.close()


def plot_lambda():
    q = np.logspace(-2, 3, 1000)

    eta = 0
    Dr = 2
    eta_AB = 1
    eta_BA = - eta_AB

    fig, axes = plt.subplots(2, 3, figsize=(9, 10), constrained_layout=True, sharex=True)
    
    # varied alpha at eta=0 and Dr=0.1
    eta = -0.3
    Dr = 3
    for alpha2 in [0.2, 0.5, 1, 1.5]:
        eta_AB = np.sqrt(alpha2)
        eta_BA = -eta_AB
        y = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q)
        axes[0, 0].plot(q, y.real, label=r"$%.1f$" % alpha2)
        axes[1, 0].plot(q, y.imag, label=r"$%.1f$" % alpha2)
    axes[0, 0].set_title("(a) $\eta=%g,D_r=%g$" % (eta, Dr), fontsize="large")
    
    # varied Dr at eta=0 and alpha=1
    eta = 0
    eta_AB = 1
    eta_BA = -eta_AB
    Dc = get_Dr_threshold(eta, eta_AB, eta_BA)
    for Dr in [1, Dc, 10]:
        y = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q)
        axes[0, 1].plot(q, y.real, label=r"$%.2f$" % Dr)
        axes[1, 1].plot(q, y.imag, label=r"$%.2f$" % Dr)
    axes[0, 1].set_title("(b) $\eta=%g,\eta_{AB}=-\eta_{BA}=%g$" % (eta, eta_AB), fontsize="large")

    # varied eta at alpha=1 and Dr=0.1
    eta_AB = 1
    eta_BA = -1
    Dr = 0.1
    for eta in [0, -0.5, -1, -2]:
        y = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q)
        axes[0, 2].plot(q, y.real, label=r"$%g$" % eta)
        axes[1, 2].plot(q, y.imag, label=r"$%g$" % eta)

    for ax in axes.flat:
        ax.axhline(0, linestyle="dashed", c="k")
        ax.axvline(np.pi * 2, linestyle="dotted", c="tab:red")
        ax.set_xscale("log")
        ax.set_xlim(q[0], q[-1])

    for ax in axes[1]:
        ax.set_xlabel(r"$q$", fontsize="x-large")
    axes[0, 2].set_title("(c) $D_r=%g,\eta_{AB}=-\eta_{BA}=%g$" % (Dr, eta_AB), fontsize="large")
    
    axes[0, 1].set_ylim(-10, 5)

    axes[0, 0].set_ylabel(r"$q\Re(\lambda_{++})$", fontsize="x-large")
    axes[1, 0].set_ylabel(r"$q\Im(\lambda_{++})$", fontsize="x-large")


    axes[0, 0].legend(title="$\eta_{AB}^2=$", fontsize="large")
    axes[0, 1].legend(title="$D_r=$", fontsize="large")
    axes[0, 2].legend(title="$\eta=$", fontsize="large")

    plt.show()
    # plt.savefig("Re_lambda.pdf")
    plt.close()

plot_lambda()
```

```python
def plot_lambda_w_Dt():
    q = np.logspace(-2, 3, 1000)

    eta = 0
    Dr = 0.1
    eta_AB = 0.8
    eta_BA = - eta_AB

    fig, ax = plt.subplots(1, 1, figsize=(5, 6), constrained_layout=True)
    

    y0 = cal_lambda_pp(eta, eta_AB, eta_BA, Dr, q).real
    for Dt in [0, 0.01]:
        y = y0 - Dt * q**2  
        ax.plot(q, y, label=r"$%g$" % Dt)

    ax.axhline(0, linestyle="dashed", c="k")
    ax.set_title("(c) $D_r=%g,\eta=%g,\eta_{AB}=-\eta_{BA}=%g$" % (Dr, eta, eta_AB), fontsize="large")

    # ax.axvline(np.pi * 2, linestyle="dotted", c="tab:red")
    ax.set_xscale("log")
    ax.set_xlim(q[0], q[-1])
    ax.set_ylim(-1, 0.5)
    ax.set_xlabel(r"$q$", fontsize="x-large")
    ax.set_ylabel(r"$\Re(\lambda_{++})$", fontsize="x-large")
    ax.legend(title=r"$\frac{D_t}{v_0}=$", title_fontsize="xx-large")
    plt.show()
    # plt.savefig("Re_lambda_Dt.pdf")
    plt.close()

plot_lambda_w_Dt()
```
