# Phase diagram in composition plane

## Linear expansion around the homogenous phase

Around the homogenous phase $(\rho_S=\phi_S, \mathbf{p}_S=0)$, the dynamics of fluctuations $\delta \rho_S$ and $\mathbf{p}_S$ are dictated by
$$
\begin{aligned}
		\partial_t \begin{pmatrix}
			\delta \rho_A \\ p_A \\ \delta \rho_B \\ p_B
		\end{pmatrix} &=
		v_0 q M
		\begin{pmatrix}
			\delta \rho_A \\ p_A \\ \delta \rho_B \\ p_B
		\end{pmatrix}
		-D_t q^2
		\begin{pmatrix}
			\delta \rho_A \\ p_A \\ \delta \rho_B \\ p_B
		\end{pmatrix},
	\end{aligned}
$$
where $M$ is a matrix depending on dimensionless parameters $\eta_{SS'}, \nu_S, \psi_S$ and $D_S$:
$$
\begin{aligned} 
	M &= \begin{pmatrix}
		0 &  -i\nu_A & 0 & 0 \\
		-\frac{i}{2} (\nu_A+ \psi_A\eta_{AA})& -D_A & -\frac{i}{2}\psi_A \eta_{AB} & 0 \\
		0 & 0 & 0 & -i \nu_B\\
		-\frac{i}{2}\psi_B \eta_{BA}& 0 & -\frac{i}{2} (\nu_B+\psi_B\eta_{BB})& -D_B
	\end{pmatrix},\\
	&= \begin{pmatrix}
		0 & -i \nu_A & 0 & 0 \\
		\varepsilon_{AA} & -D_A & \varepsilon_{AB} & 0 \\
		0 & 0 & 0 & -i \nu_B \\
		\varepsilon_{BA} & 0 & \varepsilon_{BB} & -D_B
	\end{pmatrix},
\end{aligned}
$$

where
$$
\begin{aligned}
	\nu_S &=\frac{v_{S,0}}{v_0} = \frac{\bar{v}_S}{v_0}\left[1 +
		\eta_{SA}\frac{\rho_{A,0}-\bar{\rho}_A}{\rho_0} +
		\eta_{SB}\frac{\rho_{B,0}-\bar{\rho}_B}{\rho_0}
	\right],\\
	\psi_S &=\frac{\rho_{S,0}\bar{v}_S}{\rho_0 v_0},\\
	D_S &=\frac{D_r}{v_0 q} + \frac{v_{S,0}^2}{16 v_0 D_r}q
	= \frac{D_r}{v_0 q} +\frac{\nu_S^2 v_0}{16 D_r}q,\\
	\varepsilon_{SS'}&=-\frac{i}{2}(\nu_S\delta_{S,S'}+\psi_S \eta_{SS'}).
\end{aligned}
$$

Note that the matrix $M$ has a dependence on $q$ through $D_A$ and $D_B$. To get the eigenvalues of $M$, we calculate the determinant
$$
\begin{aligned}
	|M-\lambda I| &= \left|
	\begin{matrix}
		-\lambda & -i\nu_A & 0 & 0 \\
		\varepsilon_{AA} & -D_A-\lambda & \varepsilon_{AB} & 0 \\
		0 & 0 & -\lambda & -i\nu_B \\
		\varepsilon_{BA} & 0 & \varepsilon_{BB} & -D_B - \lambda
	\end{matrix} \right|\\
	&= (D_A + \lambda)(D_B + \lambda) \lambda^2 + i \nu_B \varepsilon_{BB} (D_A+\lambda)\lambda + i \nu_A \varepsilon_{AA} (D_B+\lambda) \lambda + \nu_A\nu_B (\varepsilon_{AB}\varepsilon_{BA}-\varepsilon_{AA}\varepsilon_{BB}).
\end{aligned}
$$

The determinant can be further simplified as
$$
\begin{aligned}
|M-\lambda I| &= (D_A+\lambda)(D_B+\lambda)\lambda^2 +\beta_B(D_A+\lambda)\lambda + \beta_A (D_B+\lambda)\lambda + \alpha \beta_A\beta_B \\
&= \lambda^4 + (D_A+D_B)\lambda^3 + (\beta_A+\beta_B+D_AD_B) \lambda^2 + (\beta_B D_A + \beta_A D_B) \lambda + \alpha \beta_A \beta_B
\end{aligned}
$$
by defining
$$
\begin{aligned}
\beta_S &=i\nu_S\varepsilon_{SS}=\frac{\nu_S(\nu_S + \psi_S\eta_{SS})}{2},\\
\alpha &= 1 -\frac{\varepsilon_{AB}\varepsilon_{BA}}{\varepsilon_{AA}\varepsilon_{BB}}
\end{aligned}.
$$

## The case of $\rho_{S,0}=\bar{\rho}_S$ and $\bar{v}_A=\bar{v}_B$
We first consider the case of $\rho_{A, 0}=\bar{\rho}_A$ and $\rho_{B,0}=\bar{\rho}_B$, while keeping $\bar{v}_A=\bar{v}_B$. When $\bar{v}_A=\bar{v}_B$, without loss of generality, we can set $v_0\equiv\bar{v}_A=\bar{v}_B$.
Then $\nu_S=1$, $\psi_S=\frac{\rho_{S,0}}{\rho_0}$,
$D_S=\frac{D_r}{v_0 q}+\frac{v_0}{16D_r}q\equiv D$, and the determinant
$|M-\lambda I|$ becomes
$$
\begin{aligned}
	|M-\lambda I| &= (D+\lambda)^2 \lambda^2 + i(\varepsilon_{AA}+\varepsilon_{BB})(D+\lambda) \lambda + (\varepsilon_{AB}\varepsilon_{BA}-\varepsilon_{AA}\varepsilon_{BB}) \\
	&=\left[ (D+\lambda)\lambda +
	\frac{i(\varepsilon_{AA}+\varepsilon_{BB})}{2} 
	\right]^2 +
	\frac{(\varepsilon_{AA}-\varepsilon_{BB})^2}{4} + \varepsilon_{AB}\varepsilon_{BA}.
\end{aligned}
$$

Letting $|M-\lambda I|=0$, we get
$$
	\lambda^2 + D\lambda +
	\frac{i(\varepsilon_{AA}+\varepsilon_{BB})}{2} \pm
	\sqrt{-\frac{(\varepsilon_{AA}-\varepsilon_{BB})^2}{4} - \varepsilon_{AB}\varepsilon_{BA}} = 0,
$$
which leads to
$$
	\lambda_{\pm\pm}=\frac
	{-D\pm\sqrt{D^2-2\mu \pm 2\sqrt{\Delta}}}
	{2},
$$
where we have define
$$
\begin{aligned}
	\mu &\equiv \frac{2 + \psi_A\eta_{AA} + \psi_B\eta_{BB}}{2}, \\
	\Delta &\equiv \frac{(\psi_A\eta_{AA}-\psi_B\eta_{BB})^2}{4} + \psi_A\psi_B\eta_{AB}\eta_{BA}.
\end{aligned}
$$


Define
$$
\begin{aligned}
x&= \frac{\eta_{AA}\psi_A-\eta_{BB}\psi_B}{2}, \\
y&= \frac{\eta_{AA}\psi_A+\eta_{BB}\psi_B}{2},
\end{aligned}
$$
then
$$\begin{aligned}
\psi_A &= \frac{y+x}{\eta_{AA}}, \\
\psi_B &= \frac{y-x}{\eta_{BB}}.
\end{aligned}$$

Note that $\psi_{A,B}\geq 0$ requires
$$y\leq 0, -y\leq x \leq y.$$

Meanwhile, we have
$$\begin{aligned}
\mu &= 1+y, \\
\Delta &= (1-\alpha)x^2 + \alpha y^2,
\end{aligned}$$
where $\alpha=\frac{\eta_{AB}\eta_{BA}}{\eta_{AA}\eta_{BB}}$.

### The line of exception points

The line of exception points is $\Delta=0$. $\Delta > 0$ is equivalent to
$$
\eta_{AA}^2 \psi_A^2 + (4\eta_{AB}\eta_{BA}-2\eta_{AA}\eta_{BB})\psi_A\psi_B + \eta_{BB}^2\psi_B^2 > 0,
$$
which can be rewritten as
$$
a\psi_B^2 + b\psi_B + c > 0,
$$
with
$$
\begin{aligned}
    a&= \eta_{BB}^2, \\
    b&= (4\eta_{AB}\eta_{BA}-2\eta_{AA}\eta_{BB})\psi_A, \\
    c&= \eta_{AA}^2 \psi_A^2.
\end{aligned}
$$
We can define 
$$
\begin{aligned}
    \Delta'&\equiv b^2 - 4ac \\
    &= (4\eta_{AB}\eta_{BA}-2\eta_{AA}\eta_{BB})^2\psi_A^2 - 4\eta_{AA}^2\eta_{BB}^2\psi_A^2 \\
    &= 16\psi_{A}^2\eta_{AB}^2\eta_{BA}^2 \left(
        1 - \frac{\eta_{AA}\eta_{BB}}{\eta_{AB}\eta_{BA}}
    \right).
\end{aligned}
$$
Since $a>0$, if $\Delta'<0$, then the inequaility is true for arbitrary $\psi_B$. If $\Delta'>0$, then the inequality is true when $\psi_B < \psi_{B,-}$ or $\psi_B > \psi_{B,+}$, with
$$
\begin{aligned}
    \psi_{B,\pm}&=\frac{-b\pm\sqrt{\Delta'}}{2a} \\
    &= \frac{-(4\eta_{AB}\eta_{BA}-2\eta_{AA}\eta_{BB})\psi_A \pm 4 \psi_A|\eta_{AB}\eta_{BA}|\sqrt{1-\frac{\eta_{AA}\eta_{BB}}{\eta_{AB}\eta_{BA}}}}{2\eta_{BB}^2} \\
    &= \frac{-(4\eta_{AB}\eta_{BA}-2\eta_{AA}\eta_{BB}) \pm 4 |\eta_{AB}\eta_{BA}|\sqrt{1-\frac{\eta_{AA}\eta_{BB}}{\eta_{AB}\eta_{BA}}}}{2\eta_{BB}^2} \psi_A
\end{aligned}
$$

If $\eta_{AB}\eta_{BA} < 0$ and $\eta_{AA}\eta_{BB} > 0$, then $\Delta'>0$ is always true, and the exception lines are two straight lines $\psi_B = \psi_{B,\pm}$.

```python
def get_exception_lines(x, etaAA, etaAB, etaBA, etaBB):
    A = -(4*etaAB*etaBA-2*etaAA*etaBB)
    B = 4*np.abs(etaAB*etaBA)*np.sqrt(1-etaAA*etaBB/(etaAB*etaBA))
    C = 2 * etaBB ** 2
    y_m = (A - B)/C * x
    y_p = (A + B)/C * x
    return y_m, y_p
```

### The line of $\mu=0$
$\mu > 0$ is equivalent to $2+\eta_{AA}\psi_A+\eta_{BB}\psi_B>0$.

Meanwhile, $\mu = 0$ is equivalent to $\psi_B=-\frac{2+\eta_{AA}\psi_A}{\eta_{BB}}$.

```python
def get_line_zero_mu(x, etaAA, etaBB):
    return -(2+etaAA * x) / etaBB    
```

The line of $\mu=0$ crosses the line of $\Delta=0$ at
$$
\psi_{A,\pm}=\frac{\eta_{BB}}{2\left[
    (\eta_{AB}\eta_{BA}-\eta_{AA}\eta_{BB}) \mp |\eta_{AB}\eta_{BA}|\sqrt{1-\frac{\eta_{AA}\eta_{BB}}{\eta_{AB}\eta_{BA}}}
\right]}
$$

```python
def get_cross_point_zero_mu_zero_Delta(etaAA, etaAB, etaBA, etaBB):
    A = etaAB*etaBA - etaAA* etaBB
    B = np.abs(etaAB*etaBA)*np.sqrt(1-etaAA*etaBB/(etaAB*etaBA))
    x_p = etaBB / (A + B)
    x_m = etaBB / (A - B)
    return x_p, x_m
```

### The line of $\Delta =\mu^2$

The condition $\Delta > \mu^2$ is equivalent to
$$
(\eta_{AA}\psi_A-\eta_{BB}\psi_B)^2+4\eta_{AB}\eta_{BA} \psi_A\psi_B > (2+\eta_{AA}\psi_A+\eta_{BB}\psi_B)^2,
$$
i.e.,
$$
\eta_{AB}\eta_{BA} \psi_A\psi_B > (1+\eta_{AA}\psi_A)(1+\eta_{BB}\psi_B),
$$
which is equivalent to
$$
1+ \eta_{AA}\psi_A + (\eta_{BB}+(\eta_{AA}\eta_{BB}-\eta_{AB}\eta_{BA})\psi_A)\psi_B < 0.
$$
The line of $\Delta=\mu^2$ is just
$$
\psi_B = -\frac{1+ \eta_{AA}\psi_A}{\eta_{BB}+(\eta_{AA}\eta_{BB}-\eta_{AB}\eta_{BA})\psi_A}.
$$

```python
def get_line_delta_mu_square(x, etaAA, etaAB, etaBA, etaBB):
    return -(1+etaAA * x) / (etaBB + (etaAA*etaBB-etaAB*etaBA)*x)
```

### The line of $\frac{-\Delta}{\mu}>k$
The inequaulity $\frac{-\Delta}{\mu}>k$ is equivalent to
$$
2k(2+\psi_A\eta_{AA}+\psi_B\eta_{BB}) + (\psi_A\eta_{AA}-\psi_B\eta_{BB})^2 + 4\psi_A\psi_B\eta_{AB}\eta_{BA} < 0,
$$
i.e.,
$$
a\psi_B^2 + b\psi_B + c < 0,
$$
with
$$
\begin{aligned}
a &= \eta_{BB}^2, \\
b &= (4\eta_{AB}\eta_{BA}-2\eta_{AA}\eta_{BB})\psi_A + 2k\eta_{BB}, \\
c &= \eta_{AA}^2\psi_A^2 + 2k \eta_{AA} \psi_A + 4k.
\end{aligned}
$$
The inequality is true when and only when $\Delta'=b^2-4ac<0$ and $\psi_{B,-}< \psi_B< \psi_{B,+}$, where
$$
    \psi_{B,\pm} = \frac{-b\pm\sqrt{b^2-4ac}}{2a}.
$$

```python
def get_line_q_eq_qc(x, etaAA, etaAB, etaBA, etaBB, k):
    a = etaBB ** 2
    b = (4 * etaAB * etaBA - 2 * etaAA * etaBB) * x + 2 * k * etaBB
    c = etaAA**2 * x**2 + 2 * k * etaAA * x + 4 * k
    delta = np.sqrt(b**2 - 4 * a * c)
    y1 = (-b + delta) / (2 * a)
    y2 = (-b - delta) / (2 * a)
    return y1, y2
```

```python
def PD(etaAA, etaAB, etaBA, etaBB):
    fig, ax = plt.subplots(figsize=(6, 6))
    x = np.linspace(0, 4, 1000, endpoint=True)
    y1, y2 = get_exception_lines(x, etaAA, etaAB, etaBA, etaBB)
    line, = ax.plot(x, y1, c="tab:red")
    ax.plot(x, y2, c=line.get_c())

    # y3 = -(2 + etaAA * x) / etaBB
    y3 = get_line_zero_mu(x, etaAA, etaBB)
    ax.plot(x, y3)

    xp, xm = get_cross_point_zero_mu_zero_Delta(etaAA, etaAB, etaBA, etaBB)
    x4 = np.linspace(0, xm, 1000, endpoint=True)
    y4 = get_line_delta_mu_square(x4, etaAA, etaAB, etaBA, etaBB)
    line, = ax.plot(x4, y4)
    ax.plot(y4, x4, c=line.get_c())

    k = 0.5
    x5 = np.linspace(xm, xp, 1000, endpoint=True)

    # y5_p, y5_m = line_ratio_Delta_mu(x5, etaAA, etaAB, etaBA, etaBB, k)
    y5_p, y5_m = get_line_q_eq_qc(x5, etaAA, etaAB, etaBA, etaBB, k)
    ax.plot(x5, y5_p, c="tab:green")
    ax.plot(x5, y5_m, c="tab:green")

    ax.set_xlim(0, x[-1])
    ax.set_ylim(0, x[-1])
    plt.show()
    plt.close()

etaAA = -1
etaBB = -1
etaAB = 1
etaBA = -1

PD(etaAA, etaAB, etaBA, etaBB)
```

### In $(x, y)$ plane

1. $\Delta>0$ is equivalent to $(1-\alpha)x^2+\alpha y^2>0$. If $\alpha < 0$, the inequality is true when $x< x_-$, or $x>x_+$ with
$$
x_\pm = \pm \sqrt{\frac{-\alpha}{1-\alpha}}y.
$$
Note that $\Delta < 0$ can be true only when $\alpha< 0$. If $\alpha>1$, since $x^2<y^2$, we have $(1-\alpha)x^2+\alpha y^2< (1-\alpha)y^2 + \alpha y^2=y^2\geq 0$, so it's impossible to have negative $\Delta$ when $\alpha >1$.

2. $\mu=0$ is equivalent to $y=-1$.

3. $\Delta > \mu^2$ is equivalent to $(1-\alpha)x^2+\alpha y^2>(1+y)^2$ . If $\alpha < 0$, the inequality is true when $x<x_-$ or $x>x_+$ with
$$
x_\pm = \pm \sqrt{\frac{1+ 2y + (1-\alpha)y^2}{1-\alpha}}.
$$

4. $-\frac{\Delta}{\mu}>k$ is equivalent to $-\frac{(1-\alpha)x^2+\alpha y^2}{1+y}>k$. If $\alpha < 0$ and $\mu=1+y>0$, the inequality is true when $x\in(x_-,x_+)$, with
$$
x_\pm = \sqrt{\frac{k(1+y)+\alpha y^2}{\alpha-1}},
$$
and $y < y_-$ or $y >y_+$ with
$$
y_\pm = \frac{-k\mp \sqrt{k^2-4k\alpha}}{2\alpha}.
$$
When $k>0$ and $\alpha < 0$, we have $y_- < 0 < y_+$.
Given that $y$ need be negative to make $\psi_{A,B}$ positive, so only $y < y_-$ is requried.


```python
def plot_x_y(alpha, k=0.5):
    fig, ax = plt.subplots(figsize=(8, 4))
    y = np.linspace(-4, 0, 1000, endpoint=True)

    # psi_{A,B} >= 0
    x = y
    ax.plot(x, y, "k:")
    ax.plot(-x, y, "k:")    
    
    # Delta = 0
    x = np.sqrt(-alpha/(1-alpha)) * y
    ax.plot(x, y, c="tab:red")
    ax.plot(-x, y, c="tab:red")

    # mu = 0
    ax.axhline(-1, c="tab:blue")

    # Delta=mu**2
    x = np.sqrt((1+2*y+(1-alpha)* y**2)/(1-alpha))
    ax.plot(x, y, c="tab:orange")
    ax.plot(-x, y, c="tab:orange")

    # -Delta/mu > k
    y_max = (-k + np.sqrt(k**2 - 4 * k * alpha))/(2 * alpha)
    y_new = np.linspace(y.min(), y_max, 10000, endpoint=False)
    x = np.sqrt((k* (1+y_new) + alpha * y_new ** 2)/(alpha - 1))
    ax.plot(x, y_new, c="tab:green")
    ax.plot(-x, y_new, c="tab:green")

    ax.axis("scaled")
    ax.set_xlabel(r"$x$")
    ax.set_ylabel(r"$y$")

    ax.set_ylim(y.min(), y.max())
    ax.set_xlim(y.min(), -y.min())

    plt.show()
    plt.close()

plot_x_y(-1)
# plot_x_y(-1)
```

```python
def xy2psi(x, y, etaAA, etaBB):
    psiA = (y+x) / etaAA
    psiB = (y-x) / etaBB
    return psiA, psiB


def plot_psiA_psiB(etaAA, etaAB, etaBA, etaBB, k=0.5):
    fig, ax = plt.subplots(figsize=(6, 6))

    alpha = etaAB * etaBA / (etaAA * etaBB)
    y = np.linspace(-2, 0, 1000, endpoint=True)
    
    # Delta = 0
    x = np.sqrt(-alpha/(1-alpha)) * y

    psiA, psiB = xy2psi(x, y, etaAA, etaBB)
    ax.plot(psiA, psiB, c="tab:red")
    psiA, psiB = xy2psi(-x, y, etaAA, etaBB)
    ax.plot(psiA, psiB, c="tab:red")
    psi_max = psiB.max()

    # mu = 0
    y0 = -1
    x = np.linspace(y0, -y0, 1000, endpoint=True)
    psiA, psiB = xy2psi(x, np.ones_like(x) * y0, etaAA, etaBB)
    # print(psiA)
    # print(psiB)

    ax.plot(psiA, psiB, c="tab:blue")

    # Delta=mu**2
    x = np.sqrt((1+2*y+(1-alpha)* y**2)/(1-alpha))

    mask = y >= -1
    psiA, psiB = xy2psi(x, y, etaAA, etaBB)
    ax.plot(psiA[mask], psiB[mask], c="tab:orange")
    psiA, psiB = xy2psi(-x, y, etaAA, etaBB)
    ax.plot(psiA[mask], psiB[mask], c="tab:orange")

    # -Delta/mu > k
    y_max = (-k + np.sqrt(k**2 - 4 * k * alpha))/(2 * alpha)
    y_new = np.linspace(y.min(), y_max, 20000, endpoint=False)
    x = np.sqrt((k* (1+y_new) + alpha * y_new ** 2)/(alpha - 1))

    mask = y_new >= -1
    psiA, psiB = xy2psi(x, y_new, etaAA, etaBB)
    ax.plot(psiA[mask], psiB[mask], c="tab:green")
    psiA, psiB = xy2psi(-x, y_new, etaAA, etaBB)
    ax.plot(psiA[mask], psiB[mask], c="tab:green")

    # ax.set_xlabel(r"$x$")
    # ax.set_ylabel(r"$y$")

    ax.axis("scaled")
    ax.set_ylim(0, psi_max)
    ax.set_xlim(0, psi_max)
    ax.set_xlabel(r"$\psi_A$", fontsize="x-large")
    ax.set_ylabel(r"$\psi_B$", fontsize="x-large")

    ax.plot(20/40, 20/40, "o")
    ax.plot(30/40, 30/40, "o")

    plt.show()
    plt.close()


etaAA = etaBB = -1
etaAB = 1
etaBA = -etaAB
plot_psiA_psiB(etaAA, etaAB, etaBA, etaBB)
```
