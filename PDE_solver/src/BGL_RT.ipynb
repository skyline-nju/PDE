# BGL equations for the random torque case

Equations for the free motion of particles in the random torques case read
$$
\begin{aligned}
\dot{\mathbf{r}}_j &= v_0 \hat{\mathbf{e}} (\theta_j),\\
\dot{\theta}_j &= \zeta \gamma(\mathbf{r}) + \eta \xi(t),
\end{aligned}
$$

where $\eta$ and $\epsilon$ are respectively the strengths of the annealed noise and of the quenched disorder. $\gamma$ and $\xi$ are two white noises uniformly distributed in $[-\pi, \pi]$.

Therefore, the density of probability $f$ that a particle is in a phase space $(\mathbf{r}, \theta)$ of volume $\mathrm{d}\mathbf{r}\mathrm{d}\theta$ at time $t$ obeys

$$\partial_t f(\mathbf{r}, \theta, t)+v_0 \hat{\mathbf{e}}(\theta) \cdot \mathbf{\nabla} f(\mathbf{r}, \theta, t) + \zeta \gamma (\mathbf{r})\partial_\theta f(\mathbf{r}, \theta, t)=I_{\mathrm{source}}[f]$$

where the r.h.s. of this equation gathers the usual self-diffusion and collisional integrals modeling the effects of annealed noise and two-body interactions. The hierarchy in angular Fourier space are

$$\partial_t f_k + \frac{1}{2}(\nabla^* f_{k+1} + \nabla f_{k-1}) = [P_k - 1 + ik \zeta \gamma (\mathbf{r})]f_k + \rho_0 \sum_q J_{k, q} f_q f_{k-q},$$

where the dependencies of the modes on space and time have been dropped for simplicity, the complex gradient $\nabla = \partial_x + i\partial_y$, the angular noise moments $P_k = \exp (-k^2\eta^2 /2)$, and the mode coupling functions $J_{k, q}=P_kI_{k, q}-I_{0, q}$ with

$$
I_{k, q} = 
\begin{cases}
\frac{4}{\pi} \frac{1-(k-2q)(-1)^q \sin (\frac{k\pi}{2})}{1-(k-2q)^2} \quad &\mathrm{if}\ |k-2q| \neq 1\\
\frac{2}{\pi} \quad &\mathrm{otherwise}
\end{cases}
$$

For $k=0$,

$$\partial_t f_0 = -\frac{1}{2} (\nabla^* f_1 + \nabla f_{-1})=-\Re(\nabla^* f_1),$$

For $k=1$,

$$\partial_t f_1 + \frac{1}{2} (\nabla^* f_2 + \nabla f_0)=\mu f_1 + \rho_0 (J_{1, -1}+J_{1, 2})f^*_1 f_2,$$

where $\mu = P_1 - 1 + \rho_0 (J_{1, 0}+J_{1, 1})f_0 + i \zeta \gamma(\mathbf{r})$.

For $k=2$,

$$
f_2 = \frac{1}{2(\mu_2 + 2i\zeta 
\gamma(\mathbf{r}))} \nabla f_1 - \frac{\rho_0 J_{2, 1}}{\mu_2 + 2i \zeta\gamma(\mathbf{r})} f_1^2,
$$

where $\mu_2 = (P_2 - 1) + \rho_0 f_0 (J_{2, 0} + J_{2, 2})$ has no dependence on $\mathbf{r}$ assuming $f_0\sim 1$.

$$
\begin{aligned}
\nabla^* f_2 &= \frac{1}{2} \frac{\Delta f_1 (\mu_2 + 2i\zeta \gamma)- 2i\zeta \nabla^* \gamma\nabla f_1 }{(u_2 + 2i\zeta\gamma)^2}-\rho_0 J_{2, 1}\frac{\nabla^* f_1^2 (\mu_2 + 2i\zeta\gamma)- 2i\zeta\nabla^* \gamma f_1^2}{(\mu_2+2i\zeta\gamma)^2} \\
&=\frac{1}{2(\mu_2 + 2i\zeta \gamma)}\Delta f_1 - \frac{i\zeta}{(\mu_2+2i\zeta\gamma)^2} \nabla^* \gamma \nabla f_1-\frac{\rho_0 J_{2,1}}{\mu_2 + 2i\zeta\gamma}\nabla^* f_1^2 + \frac{2i\rho_0 J_{2, 1} \zeta}{(\mu_2 + 2i \zeta \gamma)^2}\nabla^* \gamma f_1^2
\end{aligned}
$$

Thus $f_1$ can be expressed as
$$\partial_t f_1  = -\frac{1}{2} \nabla f_0 + (\mu-\xi|f_1|^2) f_1+ \nu \Delta f_1 + \kappa_1 f_1^* \nabla f_1 + \kappa_2 \nabla^* f_1^2 + (\alpha_1 \nabla f_1 + \beta_1 f_1^2) \nabla^* \gamma,$$
where the coefficients are listed below:

| Coefficient | Value | 
| :-: | :-: |
| $\mu$ | $$P_1 - 1 + \rho_0 \rho (J_{1, 0} + J_{1, 1}) + i\zeta \gamma(\mathbf{r})$$ | 
| $\xi$ | $$\frac{\rho_0^2(J_{1, 2} + J_{1, -1})J_{2, 1}}{\mu_2+ 2i\zeta \gamma(\mathbf{r})}$$ |
| $\nu$ | $$-\frac{1}{4(\mu_2+ 2i\zeta \gamma(\mathbf{r})}$$ |
| $\kappa_1$ | $$\frac{\rho_0(J_{1, 2} + J_{1, -1})}{2(\mu_2+2i\zeta \gamma(\mathbf{r}))}$$ |
| $\kappa_2$ | $$\frac{\rho_0 J_{2, 1}}{2(\mu_2+ 2i\zeta \gamma(\mathbf{r}))}$$ |
| $\alpha_1$ | $$\frac{i\zeta}{2(\mu_2+ 2i\zeta \gamma(\mathbf{r}))^2}$$ |
| $\beta_1 $ | $$-\frac{i\zeta\rho_0 J_{2, 1}}{(\mu_2+2i\zeta \gamma (\mathbf{r}))^2}$$ |
